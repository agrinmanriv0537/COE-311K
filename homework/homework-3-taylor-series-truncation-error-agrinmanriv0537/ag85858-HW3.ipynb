```python
import numpy as np
import matplotlib.pyplot as plt

'''
AI used for nicer plotting of errors. 
Prompt: Here is my code, plot this nicely using pyplot.
Why: Nicer presentation of errors. 
'''

def taylor_sin(x, terms):

    x = np.asarray(x, dtype=float)

    term = x.copy()
    approx = term.copy()
    
    for n in range(terms - 1):
        denom = (2*n + 2) * (2*n + 3)
        term = -term * (x * x) / denom
        approx = approx + term

    return approx


x = np.arange(0, 10.0 + 0.1, 0.1)

exact = np.sin(x)

errors = {}
for N in [1, 2, 3, 4]:
    approx = taylor_sin(x, N)
    errors[N] = np.abs(approx - exact)

plt.figure()
for N in [1, 2, 3, 4]:
    plt.plot(x, errors[N], label=f"{N} term(s)")
plt.xlabel("x")
plt.ylabel("absolute error |approx - sin(x)|")
plt.title("Truncation error for Taylor approximation of sin(x)")
plt.yscale("log")
plt.grid(True)
plt.legend()
plt.show()
```
