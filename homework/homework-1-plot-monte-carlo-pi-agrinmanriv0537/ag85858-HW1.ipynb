```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math


# Monte Carlo pi approximation method
def approximatePi(numSamples):
    counter = 0
    x = 0
    y = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if ((x * x + y * y) < 1):
            counter = counter + 1
    return 4 * counter / numSamples

# Creating list to iterate over
num = 10
iterations = []
while(num <= 10000):
    iterations.append(num)
    num = num + 20

# Iterate through approximations, calculate error, and save value in list
error = []  
for i in iterations:
    approximation = approximatePi(i)
    error.append(round(abs(approximation - np.pi), 15))
    
# Plot iterations on x axis and error on y axis        
plt.plot(iterations, error)
plt.xlabel("Pi Calculation Iterations")
plt.ylabel("Error")
```
