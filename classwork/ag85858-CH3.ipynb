```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math

def approximatePi(numSamples):
    counter = 0
    x = 0
    y = 0
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if ((x * x + y * y) < 1):
            counter = counter + 1
    return 4 * counter / numSamples
```

```python
relativeError = []
trueError = []

for i in range(100):
    myPi = approximatePi(1000)
    
    relativeError.append(abs((np.pi) - myPi) / np.pi)
    trueError.append(abs(np.pi - myPi))
#     print(np.pi, myPi, relativeError, trueError)
```

```python
plt.plot(relativeError, trueError)
```

```python
plt.subplot(1,2,1)
plt.hist(relativeError)
plt.xlabel('Relative Error')
plt.subplot(1,2,2)
plt.hist(trueError)
plt.xlabel('True Error')
plt.show()
```
