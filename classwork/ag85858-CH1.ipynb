```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 6 * np.pi, 200)
```

```python
plt.plot(x, np.sin(x))
```

```python
line = [0] * int(6 * np.pi)
plt.plot(x, np.sin(x), 'g-x',
x, np.cos(x), 'c-o', line, 'k-.')
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)
size = np.random.rand(10) * 50
color = np.random.rand(10)
plt.scatter(x, y, size, color)
plt.colorbar()
plt.show()
```
